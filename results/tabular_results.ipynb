```python
import os
import glob
import pandas as pd
os.chdir("mydir")
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
#combinar todos os arquivos da lista
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
combined_csv.sort_values(['ferramenta'])
#exportar para csv
combined_csv.to_csv( "result_ferramentas.csv", index=False, encoding='utf-8-sig')
combined_csv.to_excel( "result_ferramentas.xlsx", index=False, encoding='utf-8-sig')
```
